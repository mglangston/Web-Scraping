```
# import all dependencies at the top
from time import time
from time import sleep
from random import randint
from IPython.core.display import clear_output
from warnings import warn
from bs4 import BeautifulSoup
import requests
```

```
url = 'https://www.space.com/news'
response = requests.get(url)

# make sure we got a valid response
if(response.ok):
  # get the full data from the response
  data = response.text
  soup = BeautifulSoup(data, 'html.parser')
  stories = []
  
  # define a function to process the page
def process_page(soup, stories):  
  
  # find all elements with class *content*
  raw_stories = soup.select('.content')
  stories = []

# summary = soup.find_all(class_='content')
# print(summary)

# content = soup.select('.content')
# print(content)
  for story in raw_stories:
    headline = story.select_one('.article-name').get_text()
    author = story.select_one('.by-author > span').get_text().lstrip('\n') # extract the author
    synopsis = story.select_one('.synopsis').get_text().lstrip('\n').rstrip('\n') # extract the synopsis
    date_time = story.select_one('time').attrs['data-published-date'] # extract the date and time
    story = {'headline': headline, 'author': author, 'synopsis': synopsis, 'date': date_time} # construct a dictionary
    stories.append(story) # add dictionary to list
  return stories
    
process_page(soup, stories)
    
```
